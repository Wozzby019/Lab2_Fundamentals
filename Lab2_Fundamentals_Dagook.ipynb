```python
# ==========================================
# EXERCISE 1: REDACTION PROTOCOL
# ==========================================
print("--- Exercise 1: Redaction Protocol ---")

# Setup: Replace these with your actual details
surname = "DAGOOK"
student_id = "TUPM-25-0518"
seed_num = int(student_id[8]) 
redacted_name = ""
vowels = "AEIOU"

# Logic
for char in surname:
    if char in vowels:
        redacted_name += str(seed_num)
    else:
        redacted_name += char

print(f"Original Surname: {surname}")
print(f"SEED_NUM Used: {seed_num}")
print(f"Redacted Output: {redacted_name}\n")


# ==========================================
# EXERCISE 2: FORENSIC SIGNATURE
# ==========================================
print("--- Exercise 2: Forensic Signature ---")

# Logic
# Assuming 'ENGINEER name' refers to your Surname from Task 1
first_letter = surname[0]
ascii_val = ord(first_letter)
loops = len(surname) # Total cycles to process the string

# Calculation: (ASCII * SEED_NUM) - loops
verification_hash = (ascii_val * seed_num) - loops

print(f"First Letter ASCII (A): {ascii_val}")
print(f"SEED_NUM (B): {seed_num}")
print(f"Total Loops (C): {loops}")
print(f"Verification Hash: {verification_hash}\n")


# ==========================================
# EXERCISE 3: SECURE INPUT LOOP
# ==========================================
print("--- Exercise 3: Secure Input ---")

correct_password = "DAGOOK"
attempts = 0
user_input = ""

while user_input != correct_password:
    user_input = input("Enter password to unlock: ")
    attempts += 1
    
    if user_input == correct_password:
        print("Access Granted.")
    else:
        print(f"Incorrect. This is attempt #{attempts}")

print(f"Correct Password Set: {correct_password}")
print(f"Attempts Before Success: {attempts}")
```
